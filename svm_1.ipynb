```python
import numpy as np
import pandas as pd
from pydataset import data
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn import model_selection
from sklearn.preprocessing import LabelEncoder
```

```python
#Preparing the data by removing the NA values.
df_test = pd.DataFrame(data('OFP'))
df_test == df_test.dropna()
df_test.head()
```

```python
#Sorting out the categories and the applying a label encoder to get them to a numerical input.
df_categories = df_test[['black','sex','maried','employed','privins','medicaid','region','hlth']]
df_categories = df_categories.apply(LabelEncoder().fit_transform)
```

```python
df_categories
```

```python
#Merging with the initial dataset
df_final = df_categories.merge(df_test)
```

```python
#Checking if all the values are correct and no unesscessary transformations hasnt taken place.
df_test['faminc'].isin(df_final['faminc']).value_counts()
```

```python
df_final
```

```python
#Scaling all the values for SVM
df_final = (df_final-df_final.min())/(df_final.max()-df_final.min())
```

```python
df_final
```

```python
#Sperating independent and dependent values
X = df_final[['black','sex','employed','privins','medicaid','region','hlth','ofp','ofnp','opp','opnp','emr','hosp',	'numchron',	'adldiff', 'age',	'school',	'faminc']]
Y = df_final[['maried']]
```

```python
#Seperating into x_test and y_test
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y,test_size=.3,random_state=1)
```

```python
#Constructing the model
model = svm.LinearSVC(C=1)
```

```python
#Fitting the data into the model
model.fit(X_train,Y_train)
```

```python
#Printing the model score
model.score(X_train,Y_train)
```

```python
#Predicting for the new values
y_pred = model.predict(X_test)
```

```python
#Printing a classification reports
print(classification_report(Y_test,y_pred))
```
